```python
#pip install pandas_datareader
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

# Banking Sector

```python
berkshire = pd.read_csv("BRK-A.csv")
JpMorgan = pd.read_csv("JPM.csv")
visa = pd.read_csv("V.csv")
bankOfamerica = pd.read_csv("BAC.csv")
wellsFargo = pd.read_csv("WFC.csv")
berkshire.head()
```

```python
stocks = pd.DataFrame({"BRK": berkshire["Adj Close"],
                      "JPM": JpMorgan["Adj Close"],
                      "Visa": visa["Adj Close"],
                      "BAC": bankOfamerica["Adj Close"],
                       "WFC": wellsFargo["Adj Close"]})
stock_return = stocks.apply(lambda x: x / x[0])
stock_return.plot(grid = True).axhline(y = 1, color = "black", lw = 2)
plt.show()
```

In the above graph we can clearly see that there are only two stocks which did not drop below the line and have been returning positive growth with some fluctuations. The purple line is for Wells Fargo and Green line is for Visa. Therefore, these two stocks are the safest bets. 

```python
stock_change = stocks.apply(lambda x: np.log(x) - np.log(x.shift(1)))
stock_change = stock_change.dropna()
stock_change.plot(grid = True).axhline(y = 0, color = "black", lw = 2)
plt.show()
```

From the above graph we can see the volatality of the stocks. We can see the stocks for Bank of America and Jp Morgan Chase are the most volatile ones and Wells Fargo is less volatile comaparitively.

```python
v = [berkshire,JpMorgan,visa,bankOfamerica,wellsFargo]
for i in v:
    i['Daily Return'] = i['Adj Close'].pct_change()

k = 0
for i in v:
        sns.kdeplot(i['Daily Return'].dropna(),
            color = 'blue', label = stock_return.columns.values[k])
        k=k+1
        plt.legend()
        plt.show()
```

Another noticeable point from the above graphs is that it is in shape of a bell curve. The reason for this is because most people tend to buy the stocks as soon as the market opens and decide to sell it before the market is closed.

# Healthcare Sector

```python
JNJ = pd.read_csv("JNJ.csv")
MRK = pd.read_csv("MRK.csv")
PFE = pd.read_csv("PFE.csv")
NVS = pd.read_csv("NVS.csv")
UNH = pd.read_csv("UNH.csv")
```

```python
UNH.head()
```

```python
stocks_health = pd.DataFrame({"JNJ": JNJ["Adj Close"],
                      "PFE": PFE["Adj Close"],
                      "NVS": NVS["Adj Close"],
                      "UNH": UNH["Adj Close"],
                       "MRK": MRK["Adj Close"]})
stock_return_health = stocks_health.apply(lambda x: x / x[0])
stock_return_health = stock_return_health.dropna()
stock_return_health.plot(grid = True).axhline(y = 1, color = "black", lw = 2)
plt.show()
```

In the above graph we can see only PFE managed to return profits while the other showed lower growth levels or negative growth. MRK is still relatively a safe bet because it dipped for the least amount of time and managed to growth better than PFE. UNH also displayed a drop but also faired good growth.

```python
stock_change_health = stocks_health.apply(lambda x: np.log(x) - np.log(x.shift(1)))
stock_return_health = stock_change_health.dropna()
stock_change_health.plot(grid = True).axhline(y = 0, color = "black", lw = 2)
plt.show()
```

```python
v1 = [MRK,JNJ,NVS,PFE,UNH]
for i in v1:
    i['Daily Return'] = i['Adj Close'].pct_change()
k = 0
for i in v1:
        sns.kdeplot(i['Daily Return'].dropna(), 
            color = 'blue', label = stock_return_health.columns.values[k])
        k=k+1
        plt.legend()
        plt.show() 
```
